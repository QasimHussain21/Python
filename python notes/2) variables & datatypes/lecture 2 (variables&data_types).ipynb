# Data Types in Python

```python
a = 1
b = "qasim"
c = True
d = 1.2
e = 1+2j
f = (1,2,3)
g = [10,20,30]
h = {10,20,30}
i = {'a' : 20, 'b' : 30}
# print(a,b,c,d,e)
print(a)
print(b)
print(c)
print(d)
print(e)
print(f)
print(g)
print(h)
print(i)
print(type(a),type(b),type(c),type(d),type(e),type(f),type(g),type(h),type(i))
```

```python
num1 =[1,2,3,4]
num2 ={1,2,3,4}
num3 =(1,2,3,4)
print(num1, num2, num3)
print(type(num1),type(num2),type(num3))
```
